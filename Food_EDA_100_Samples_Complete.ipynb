```python
# Set random seed for reproducibilitynp.random.seed(42)# Define global parameterscountries = ['Italy', 'India', 'China', 'Mexico', 'Japan', 'France', 'Thailand', 'USA', 'Spain', 'Greece']food_categories = ['Appetizer', 'Main Course', 'Dessert', 'Beverage', 'Snack']ingredients_pool = [    'Tomato', 'Garlic', 'Onion', 'Rice', 'Pasta', 'Chicken', 'Beef', 'Fish',     'Cheese', 'Milk', 'Eggs', 'Butter', 'Olive Oil', 'Soy Sauce', 'Ginger',    'Chili', 'Coconut', 'Lemon', 'Basil', 'Cilantro', 'Cumin', 'Turmeric',    'Pepper', 'Salt', 'Sugar', 'Flour', 'Yogurt', 'Cream', 'Honey', 'Vinegar']print("="*80)print(" GENERATING COMPREHENSIVE FOOD DATASETS")print("="*80)# ============================================================================# DATASET 1: MAIN FOOD NUTRITION DATABASE# ============================================================================n_samples = 500food_data = pd.DataFrame({    'food_id': range(1, n_samples + 1),    'country': np.random.choice(countries, n_samples),    'category': np.random.choice(food_categories, n_samples),    'calories': np.random.randint(50, 800, n_samples),    'protein_g': np.random.uniform(1, 50, n_samples).round(2),    'carbs_g': np.random.uniform(5, 100, n_samples).round(2),    'fat_g': np.random.uniform(0.5, 40, n_samples).round(2),    'fiber_g': np.random.uniform(0, 15, n_samples).round(2),    'sugar_g': np.random.uniform(0, 50, n_samples).round(2),    'sodium_mg': np.random.randint(0, 2000, n_samples),    'price_usd': np.random.uniform(2, 50, n_samples).round(2),    'prep_time_min': np.random.randint(10, 180, n_samples),    'rating': np.random.uniform(3.0, 5.0, n_samples).round(1),    'servings': np.random.randint(1, 8, n_samples),    'is_vegetarian': np.random.choice([True, False], n_samples, p=[0.3, 0.7]),    'is_vegan': np.random.choice([True, False], n_samples, p=[0.15, 0.85]),    'is_gluten_free': np.random.choice([True, False], n_samples, p=[0.25, 0.75])})print(f"✓ Dataset 1 - Food Nutrition: {food_data.shape[0]} rows, {food_data.shape[1]} columns")# ============================================================================# DATASET 2: RECIPES WITH INGREDIENTS# ============================================================================recipe_names = [    'Pasta Carbonara', 'Chicken Tikka Masala', 'Kung Pao Chicken', 'Tacos al Pastor',    'Sushi Roll', 'Beef Bourguignon', 'Pad Thai', 'Classic Hamburger', 'Paella Valenciana',    'Greek Salad', 'Pizza Margherita', 'Butter Chicken', 'Fried Rice', 'Quesadilla',    'Tonkotsu Ramen', 'French Croissant', 'Green Curry', 'BBQ Ribs', 'Tortilla Española',    'Moussaka', 'Tom Yum Soup', 'Falafel', 'Biryani', 'Pho', 'Lasagna']recipes_data = []for i, recipe in enumerate(recipe_names):    num_ingredients = np.random.randint(5, 15)    ingredients_list = np.random.choice(ingredients_pool, num_ingredients, replace=False)        recipes_data.append({        'recipe_id': i + 1,        'recipe_name': recipe,        'country': np.random.choice(countries),        'ingredients': ', '.join(ingredients_list),        'num_ingredients': num_ingredients,        'difficulty': np.random.choice(['Easy', 'Medium', 'Hard'], p=[0.3, 0.5, 0.2]),        'cooking_time_min': np.random.randint(15, 180),        'popularity_score': np.random.uniform(70, 100).round(1),        'views': np.random.randint(1000, 100000),        'likes': np.random.randint(100, 50000)    })recipes_df = pd.DataFrame(recipes_data)print(f"✓ Dataset 2 - Recipes: {recipes_df.shape[0]} rows, {recipes_df.shape[1]} columns")# ============================================================================# DATASET 3: BEVERAGE CONSUMPTION BY COUNTRY# ============================================================================beverages = ['Coffee', 'Tea', 'Wine', 'Beer', 'Juice', 'Soda', 'Water', 'Milk', 'Spirits']beverage_data = []for country in countries:    for beverage in beverages:        consumption = np.random.uniform(10, 200).round(1)        beverage_data.append({            'country': country,            'beverage': beverage,            'consumption_liters_per_capita': consumption,            'avg_price_usd': np.random.uniform(0.5, 20).round(2),            'market_share_percent': np.random.uniform(3, 25).round(1),            'growth_rate_percent': np.random.uniform(-5, 15).round(1),            'production_million_liters': consumption * np.random.uniform(50, 500)        })beverage_df = pd.DataFrame(beverage_data)print(f"✓ Dataset 3 - Beverages: {beverage_df.shape[0]} rows, {beverage_df.shape[1]} columns")# ============================================================================# DATASET 4: SEASONAL INGREDIENTS# ============================================================================seasons = ['Spring', 'Summer', 'Fall', 'Winter']seasonal_ingredients_dict = {    'Spring': ['Asparagus', 'Peas', 'Strawberries', 'Artichokes', 'Spinach'],    'Summer': ['Tomato', 'Corn', 'Watermelon', 'Peaches', 'Zucchini'],    'Fall': ['Pumpkin', 'Apples', 'Squash', 'Grapes', 'Mushrooms'],    'Winter': ['Cabbage', 'Oranges', 'Kale', 'Potatoes', 'Brussels Sprouts']}seasonal_data = []for season, ingredients in seasonal_ingredients_dict.items():    for ingredient in ingredients:        seasonal_data.append({            'season': season,            'ingredient': ingredient,            'availability_score': np.random.uniform(70, 100).round(1),            'price_per_kg_usd': np.random.uniform(1, 25).round(2),            'vitamin_content': np.random.choice(['High', 'Medium', 'Low']),            'shelf_life_days': np.random.randint(3, 30),            'organic_premium_percent': np.random.uniform(20, 80).round(1)        })seasonal_df = pd.DataFrame(seasonal_data)print(f"✓ Dataset 4 - Seasonal Ingredients: {seasonal_df.shape[0]} rows, {seasonal_df.shape[1]} columns")# ============================================================================# DATASET 5: TIME SERIES - CUISINE POPULARITY# ============================================================================dates = pd.date_range('2020-01-01', '2024-12-31', freq='M')cuisines = ['Italian', 'Chinese', 'Indian', 'Mexican', 'Japanese', 'French', 'Thai']time_series_data = []for date in dates:    for cuisine in cuisines:        base_popularity = np.random.uniform(65, 85)        trend = (date.year - 2020) * 2.5        seasonal_effect = 8 * np.sin(2 * np.pi * date.month / 12)        noise = np.random.normal(0, 3)        popularity = base_popularity + trend + seasonal_effect + noise                time_series_data.append({            'date': date,            'cuisine': cuisine,            'popularity_score': round(popularity, 2),            'searches': np.random.randint(5000, 100000),            'restaurant_count': np.random.randint(100, 5000),            'avg_rating': np.random.uniform(3.5, 4.8).round(1)        })time_series_df = pd.DataFrame(time_series_data)print(f"✓ Dataset 5 - Time Series: {time_series_df.shape[0]} rows, {time_series_df.shape[1]} columns")# ============================================================================# DATASET 6: DETAILED NUTRITIONAL COMPARISON# ============================================================================comparison_foods = [    'Quinoa', 'Brown Rice', 'White Rice', 'Couscous', 'Bulgur Wheat',    'Wild Salmon', 'Chicken Breast', 'Tofu', 'Lean Beef', 'Red Lentils',    'Greek Yogurt', 'Almonds', 'Avocado', 'Sweet Potato', 'Broccoli']nutrition_comparison = []for food in comparison_foods:    nutrition_comparison.append({        'food_item': food,        'calories_per_100g': np.random.randint(80, 600),        'protein_g': np.random.uniform(2, 35).round(1),        'carbs_g': np.random.uniform(5, 75).round(1),        'fat_g': np.random.uniform(0.5, 55).round(1),        'fiber_g': np.random.uniform(0.5, 15).round(1),        'iron_mg': np.random.uniform(0.5, 12).round(2),        'calcium_mg': np.random.uniform(10, 400).round(1),        'vitamin_c_mg': np.random.uniform(0, 90).round(1),        'vitamin_d_mcg': np.random.uniform(0, 20).round(1),        'omega3_g': np.random.uniform(0, 3).round(2)    })nutrition_df = pd.DataFrame(nutrition_comparison)print(f"✓ Dataset 6 - Nutrition Comparison: {nutrition_df.shape[0]} rows, {nutrition_df.shape[1]} columns")# Summaryprint("\n" + "="*80)print(" ✅ ALL DATASETS GENERATED SUCCESSFULLY!")print("="*80)print(f"\nTotal datasets: 6")print(f"Total records: {sum([len(df) for df in [food_data, recipes_df, beverage_df, seasonal_df, time_series_df, nutrition_df]])}")print(f"Total memory: ~{sum([df.memory_usage(deep=True).sum() for df in [food_data, recipes_df, beverage_df, seasonal_df, time_series_df, nutrition_df]]) / 1024 / 1024:.2f} MB")print("\n🎉 Ready for analysis!")
```
